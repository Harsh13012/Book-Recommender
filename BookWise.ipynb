```python
##Popularity Based Recommendation System
```

```python
ratings_df = ratings.merge(books,on="ISBN")
```

```python
ratings_df
```

```python
num_rating_df=ratings_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
avg_rating_df = ratings_df.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
filtered_df = popular_df[popular_df['num_ratings'] >= 250]
```

```python
filtered_df=filtered_df.sort_values('avg_rating',ascending=False).head(50)
```

```python
filtered_df=filtered_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
filtered_df
```

```python
##Collaborative Filtering Based Recommendation System
```

```python
temp_df = ratings_df.groupby('User-ID').count()['Book-Rating']>200
```

```python
temp_df[temp_df]
```

```python
users_index = temp_df[temp_df].index
```

```python
filtered_rating = ratings_df[ratings_df['User-ID'].isin(users_index)]
```

```python
temp_df = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = temp_df[temp_df ].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pvt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pvt.fillna(0,inplace=True)
```

```python
pvt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(pvt)
```

```python
def recommend(book_name):
    index = np.where(pvt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similar[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_books:
        item=[]
        temp_df = books[books['Book-Title']==pvt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('Message in a Bottle')
```
